```python
from sqlalchemy import create_engine
import pandas as pd
```

### DBからデータを取り出してcsvに出力してみる

```python
db_path = '/home/tadatsugu/python-netkaiba-scraping/src/keiba.db'
tables = [
    "horse_results",
    "race_entries",
    "races",
    "payouts",
]

for table in tables:
    sql = f"select * from {table} limit 5"

    # SQLAlchemyのエンジンを作成
    engine = create_engine(f"sqlite:///{db_path}")
    # SQLクエリを実行してデータフレームに読み込む
    df = pd.read_sql_query(sql, engine)
    df.to_csv(f'{table}.csv', index=False)
```

## 関数

```python
def load_data_from_db(table_name):
    '''
    DBからデータを読み込む関数
    '''
    query = f"SELECT * FROM {table_name}"
    return pd.read_sql(query, engine)
```

### DBに接続してテーブルをDataFrameとして読み込む

```python
import pandas as pd
from sqlalchemy import create_engine

db_path = '/home/tadatsugu/python-netkaiba-scraping/src/keiba.db'
engine = create_engine(f"sqlite:///{db_path}")

tables = ["horse_results", "race_entries", "races", "payouts"]

# データを格納する辞書
dfs = {}

# テーブル名をループで回す
for table in tables:
    # SQLクエリで読み込み
    dfs[table] = load_data_from_db(table)
```

```python
for name, df in dfs.items():
    print(f"\nテーブル名：{name}")
    print(df.dtypes)
```
